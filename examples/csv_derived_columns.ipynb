```python
import pandas as pd
import shlex
import operator

def scrub( input ):
    try:
        return( int(input) )
    except ValueError:
        try:
            return( float(input) )
        except ValueError:
            return input

operators = {
    '+':operator.add,
    '-':operator.sub,
    '*':operator.mul,
    '/':operator.truediv,
    '//':operator.floordiv
}

with open('csv_sample.csv','r') as input_file:
    master = pd.read_csv(input_file)

rules = []
with open('derived_columns.rules','r') as rules_file:
    for line in rules_file.readlines():
        if line[0] == "#":
            pass
        else:
            pieces = shlex.split(line)
            # two lines for readability:  we want the column if it's a string, otherwise the value
            operands = [ scrub(pieces[i]) for i in [2,4] ]
            operands = [ master[ i ] if isinstance(i, str) else i for i in operands ]

            op = operators[ pieces[3] ]
            
            master[ pieces[0] ] = op(operands[0],operands[1])
```
