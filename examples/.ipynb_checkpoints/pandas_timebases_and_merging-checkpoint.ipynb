```python
import pandas as pd
master = pd.DataFrame( { "TimeStamp": ['2020-01-01 01:01:02', '2020-01-01 01:01:04', '2020-01-01 01:01:06', '2020-01-01 01:01:08'], "a":[2,4,6,8] } )
master[ "TimeStamp" ] = pd.to_datetime(master[ "TimeStamp" ])

a = pd.DataFrame( { "TimeStamp": ['2020-01-01 01:01:01', '2020-01-01 01:01:02', '2020-01-01 01:01:03', '2020-01-01 01:01:04', '2020-01-01 01:01:05'], "Data":[11,12,13,14,15] } )
b = pd.DataFrame( { "TimeStamp": ['2020-01-01 01:01:06', '2020-01-01 01:01:07', '2020-01-01 01:01:08', '2020-01-01 01:01:09', '2020-01-01 01:01:10'], "Data":[16,17,18,19,20] } )
a[ "TimeStamp" ] = pd.to_datetime(a[ "TimeStamp" ])
b[ "TimeStamp" ] = pd.to_datetime(b[ "TimeStamp" ])
```

```python
### WIN, but depends on order ###
pd.concat( [a,b] )
```

```python
### FAIL ###
pd.concat( [b,a] )
```

```python
### WIN ###
pd.merge_ordered( a,b )
```

```python
### WIN ###
pd.merge_ordered( b,a )
```

```python
### FAIL ###
pd.merge_ordered( a,b, on="TimeStamp")
```

```python
### WIN, but not worth the extra effort? ###
pd.merge_ordered( a,b, on=["TimeStamp","Data"] )
```

```python
### NOT BAD YET ###
pd.merge_asof( master,a )
```

```python
### FAIL ###
pd.merge_asof( pd.merge_asof( master,a ), b )
```

```python
### FAIL ###
temp=pd.merge_asof( master,a )
pd.merge_asof( temp,b )
```

```python
### FAIL ###
pd.merge_asof( temp,b, on="TimeStamp" )
```

```python
### FAIL, asof does not work like normal merge ###
pd.merge_asof( temp,b, on=["TimeStamp","Data"] )
```

```python
### WIN, and what needs to be in the final implementation ###
temp = pd.merge_ordered( a,b )
pd.merge_asof( master, temp )
```

```python
# Now let's explore missing data
# Defaulting to backwards:
# OK if data is front-aligned
pd.merge_asof( master,a )
```

```python
# But not great if data starts after clock:
pd.merge_asof( master, b )
```

```python
# Much Better
pd.merge_asof( master, b, direction='nearest')
```

```python
# Much Better
pd.merge_asof( master, a, direction='nearest')
```

```python
# Doesn't affect big picture (in this example at least)
temp = pd.merge_ordered( a,b )
pd.merge_asof( master, temp, direction='nearest' )
```
